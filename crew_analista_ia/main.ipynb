```python
from dotenv import load_dotenv
from crewai import Agent, Task, Crew, Process
from crewai_tools import SerperDevTool, ScrapeWebsiteTool

# Carregar vari √° veis de ambiente do arquivo .env
load_dotenv ()
# Ferramentas de pesquisa e scraping
search_tool = SerperDevTool ()
scrape_tool = ScrapeWebsiteTool ()
# Inicializar o modelo OpenAI
gpt_mini = 'gpt-4o-mini'
gpt4o = 'gpt-4o'
site_url = ''
```

```python
agente_de_scrap_config = {
    "role":"Extrator de Informa√ß√µes de Sites",
    "backstory":f"""Voc√™ √© um especialista em scraping de sites e coleta de informa√ß√µes relevantes.""",
    "goal":f"""Celetar informa√ß√µes detalhadas sobre os servi√ßoes, produtos e opera√ß√µes da empresa a partir do site {site_url}. As informa√ß√µes extra√≠das ser√£o a base para que o agente de solu√ß√µes de IA avalie os dados do site. Identifique o 'core business' ou 'core product' da empresa. Isso √© muito importante para realizar uma an√°lise personalizada em outras etapas.""",
    "verbose":True,
    "memory":True,
    "llm":gpt_mini,
    "tools":[scrape_tool]
}

agente_solucoes_ia_config = {
    "role": "Pesquisador de Or√ßamentos para Cria√ß√£o de Sites",
    "backstory": f"""Voc√™ √© um especialista em an√°lise de or√ßamentos para cria√ß√£o de sites, capaz de encontrar as melhores op√ß√µes para empresas de diferentes portes e nichos.""",
    "goal": f"""Voc√™ receber√° dados extra√≠dos do site {site_url} e o 'core business' ou 'core product' identificado. Com base nessas informa√ß√µes, voc√™ deve pesquisar or√ßamentos para cria√ß√£o de sites que sejam adequados √†s necessidades da empresa, considerando solu√ß√µes econ√¥micas e eficazes para o desenvolvimento do site. Leve em considera√ß√£o o 'core business' ou 'core product' da empresa e busque or√ßamentos que atendam a esses requisitos de forma otimizada.""",
    "verbose": True,
    "memory": True,
    "llm": gpt_mini,
    "tools": [search_tool]
}

agente_analista_swot_config = {
    "role":"Analista e consultor de SWOT",
    "backstory":f"""Voc√™ √© um especialista em an√°lise estrat√©gica, utilizando a matriz SWOT para fornecer uma avalia√ß√£o abrangente das for√ßas, fraquezas, oportunidades e amea√ßas de uma organiza√ß√£o. Este m√©todo anal√≠tio permite identificar e alavancar pontos fortes internos, mitigar fraquezas, explorar oportunidades de mercado e prever amea√ßas externas. Ao aplicar a matriz SWOT, voc√™ ajuda empresas a desenvolver estrat√©gias eficazes que maximizam seu potencial competitivo, aprimoram a tomada de decis√µes e promovem um crescimento sustent√°vel. Sua abordagem detalhada e orientada por dados garante que cada an√°lise seja adaptada √†s necessidades espec√≠ficas do cliente, resultando em planos de a√ß√£o precisos e impactantes.""",
    "goal":f"""Analisar as informa√ß√µes coletadas e fornecer an√°lises estrat√©gicas detalhadas e precisas para ajudar a empresa a identificar e alavancar suas for√ßas, mitigar fraquezas, explorar oportunidades de mercado e prever amea√ßas externas.""",
    "verbose":True,
    "memory":True,
    "llm":gpt_mini,
    "tools":[search_tool]
}

agente_de_financiamento_config = {
    "role":"Analista de Financiamento empresarial",
    "backstory":f"""Voc√™ √© um especialista de financimento empresarial, encontrando formas inovadoras e eficazes de captar recursos para empresas. Com um profundo conhecimento das op√µes de financiamento dispon√≠veis (privados ou governamentais), desde empr√©stimos tradicionais at√© investimento de capital de risco e crowdfunding, voc√™ ajuda as empresas a identificar e acessar as melhores fontes de capital para suas necessidades espec√≠ficas. Sua habilidades em negociar termos favor√°veis e sua perspic√°cia em elaborar propostas atraentes garantem que os neg√≥cios obtenham o financiamento necess√°rio para crescer e prosperar. Com sua orienta√ß√£o, as empresas conseguem n√£o apenas assegurar recursos financeiro, mas tamb√©m fortalecer sua posi√ß√£o no mercado e atingir seus objeticos estrat√©gicos.""",
    "goal":f"""Procurar na web por oportunidades reais de financiameno da ado√ß√£o de IA pela empresa. Identificar e alavancar a implanta√ß√£o de intelig√™cia artificial na empresa. N√£o seja gen√©rico nas sugest√µes, quero uma an√°lise personalizada e real, com poss√≠veis valores e institui√ß√µes de financiamento.""",
    "verbose":True,
    "memory":True,
    "llm":gpt_mini,
    "tools":[search_tool]
}

agent_analise_recomendacao_config = {
    "role":"Consolidar e Analisar Dados de IA",
    "backstory":f"""Voc√™ √© um consultor experiente em tecnologias de IA  e tem um hist√≥rico comprovado de ajudar empresas a implementar solu√ß√µes tecnol√≥gicas. Sua habilidade em consolidar informa√ß√µes de m√∫ltiplas fontes garante que voc√™ possa fornecer uma vis√£o abrangente e detalhada.""",
    "goal":f"""Consolidar todas as informa√ß√µes extra√≠das, pesquisadas e analisadas pelos agentes anteriores. Crie um relat√≥rio final detalhado em formado markdown sobre como integrar IA de maneira eficiente e econ√¥mica na empresa em quest√£o. Leve sempre em considera√ß√£o o 'core business' ou 'core product' identificado para gerar uma an√°lise personalizada para a empresa. N√£o seja gen√©rico! Especialize o texto da an√°lise com base em todas as informa√ß√µes dispon√≠veis anteriormente. Gere um texto com uma combina√ß√£o equilibrada de par√°grafos cont√≠nuos, subt√≥picos e listas (quando necess√°rio) e n√£o se preocupe com o consumo dos tokens de sa√≠da. Quero uma an√°lise detalhada.""",
    "verbose":True,
    "memory":True,
    "llm":gpt4o,
}

scraper = Agent (config = agente_de_scrap_config)
analista_de_ia = Agent (config = agente_solucoes_ia_config)
analista_swot = Agent (config = agente_analista_swot_config)
analista_financiamento = Agent (config = agente_de_financiamento_config)
analista_recomendacao = Agent (config = agent_analise_recomendacao_config)
```

```python
extrair_informacoes_site_config = {
    "description":"Extrair informa√ß√µes detalhadas sobre os servi√ßos, produtos e opera√ß√µes da empresa a partir do site {site_url}.",
    "expected_output":"Um relat√≥rio detalhado contendo as informa√ß√µes sobre servi√ßos, produtos e opera√ß√µes da empresa.",
    "tools":[scrape_tool],
    "agent":scraper
}

pesquisar_solucoes_ia_config = {
    "description":"Pesquisar solu√ß√µes de IA econ√¥micas que podem ser integradas nos processos da empresa. Considerar ferramentas e tecnologias que sejam acess√≠veis e eficientes.",
    "expected_output":"Uma lista de solu√ß√µes de IA com detalhes sobre custos, benef√≠cios e poss√≠veis aplica√ß√µes.",
    "tools":[search_tool],
    "agent":analista_de_ia
}

relatorio_swot_config = {
    "description":"Analisar as informa√ß√µes coletadas e fornecer an√°lises estrat√©gicas detalhadas e precisas utilizando a matriz SWOT. Identificar for√ßas, fraquezas, oportunidades e amea√ßas da empresa com base nas informa√ß√µes coletadas.",
    "expected_output":"Um relat√≥rio detalhado com a an√°lise SWOT da empresa, incluindo for√ßas, fraquezas, oportunidades e amea√ßas.",
    "tools":[search_tool],
    "agent":analista_swot
}

relatorio_financiamento_estrategico_config = {
    "description":f"""Realizar uma an√°lise detalhada das op√ß√µes de financiamento poss√≠veis para a empresa. Investigar diversas fontes de capital poss√≠veis na web para a empresa financiar a ado√ß√£o de inteligencia artificial em seus processos. Analisar os termos, benef√≠cios e riscos associados a cada op√ß√£o e recomendar as melhores estrat√©gias para captar recursos""",
    "expected_output":f"""Uma an√°lise detalhada com uma compara√ß√£o as op√µes de financiamento destacando as melhores estrat√©gias para a empresa captar recursos. A an√°lise deve conter listas e tabelas comparativas que sirva de base para a escrita do relat√≥rio final.""",
    "agent":analista_financiamento
}

relatorio_final_config = {
    "description":f"""Gerar um relat√≥rio final e consolidar todas as informa√ß√µes extra√≠das, pesquisadas e analisadas pelos agentes anteriores. Fornecer recomenda√ß√µes detalhadas sobre como integrar IA de maneira econ√¥mica na empresa. O relat√≥rio deve incluir um plano de a√ß√£o e sugest√µes espec√≠ficas.""",
    "expected_output":f"""Um relat√≥rio detalhado com recomenda√ß√µes sobre:
    1 - Solu√ß√µs de integra√ß√£o de IA na empresa;
    2 - An√°lise SWOT;
    3 - Plano de a√ß√£o;
    4 - Possibilidades de financiamento;
    N√£o seja gen√©rico. Quero n√∫meros e reais possibilidades.
    5 - Sugest√µes espec√≠ficas.
    
    A sa√≠da deve ser em Markdown. √â muito importante que o texto n√£o contenha coisas como " '''markdown " e " '''". """,
    "output_file":"analise.md",
    "agent":analista_recomendacao
}

extrair_informacoes_site = Task (config = extrair_informacoes_site_config)
pesquisar_solucoes_ia = Task (config = pesquisar_solucoes_ia_config)
relatorio_swot = Task (config = relatorio_swot_config)
relatorio_financiamento_estrategico = Task (config = relatorio_financiamento_estrategico_config)
relatorio_final = Task (config = relatorio_final_config)
```

```python
crew = Crew (
    agents = [scraper, analista_de_ia, analista_swot, analista_financiamento, analista_recomendacao],
    tasks = [extrair_informacoes_site, pesquisar_solucoes_ia, relatorio_swot, relatorio_financiamento_estrategico, relatorio_final],
    process = Process.sequential
)
```

```python
site = 'https://unifacisa.edu.br/'
result = crew.kickoff(inputs={'site_url':site})

print(result)
```
